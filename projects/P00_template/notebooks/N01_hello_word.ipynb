```python
import os

from llm_inference import backends, cache

llm_backend = backends.MistralAsyncBackend(
    api_key=os.environ["MISTRAL_API_KEY"],
    cache_storage=cache.DiskCacheStorage("llm")
)

response = await llm_backend.infer_one(
    prompt="Hello word! in JSON format",
    model_config={
        "model": "mistral-small-latest",
        "temperature": 1.0,
        "max_tokens": 50,
        "random_seed": 42,
        "response_format": {
            "type": "json_object",
        }
    }
)

print(llm_backend._parse_response(response))
```
